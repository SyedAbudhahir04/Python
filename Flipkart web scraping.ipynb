```python
#creating a function to automate 

def check_price():
    from bs4 import BeautifulSoup
    import requests
    import time
    import datetime
    
    url = ('https://www.flipkart.com/luxor-premium-exercise-regular-notebook-ruled-250-pages/p/itma7b414125e050?pid=DIAFHHRDGGG8TG73&lid=LSTDIAFHHRDGGG8TG73H7ORMH&marketplace=FLIPKART&q=luxor+notebook&store=dgv%2Frxj%2Fazx&spotlightTagId=BestsellerId_dgv%2Frxj%2Fazx&srno=s_1_2&otracker=AS_QueryStore_OrganicAutoSuggest_5_5_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_5_5_na_na_na&fm=search-autosuggest&iid=40ebe648-111a-4ad2-89d6-967f3ebd8029.DIAFHHRDGGG8TG73.SEARCH&ppt=sp&ppn=sp&ssid=14csw4t3y80000001709721446932&qH=4faccd7654814963')

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36"}

    page = requests.get(url,headers=headers)
    
    soup1 = BeautifulSoup(page.content,'html.parser')
    soup2 = BeautifulSoup(soup1.prettify(),'html.parser')
    
    title = soup2.find('span', class_ = 'B_NuCI').get_text()[0:70].strip()
    price = soup2.find('div', class_ = '_30jeq3 _16Jk6d').get_text().strip()[1:]
    
    import datetime
    today = datetime.date.today()
    
    import csv
    header = ['Title','Price','Date']
    body = [title,price,today]
    # Run this only once, else header will be created again and again
    #with open('FlipkartWebScrap','w',newline='',encoding = 'UTF8') as f:
    #    writer = csv.writer(f)
    #    writer.writerow(header)
    #    writer.writerow(body)
    # this is to append data each time while the fuc runs
    with open('FlipkartWebScrap','a+',newline='',encoding = 'UTF8') as f:
        writer = csv.writer(f)
        #writer.writerow(header)
        writer.writerow(body)
        
```
